```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
id_series = pd.Series(['ID' + str(i).zfill(3) for i in range(800)])
```

```python
groups = np.random.choice(['A', 'B', 'C'], size=800)
```

```python
feature_1 = np.random.normal(2500, (2800-2200)/2, size=800)
feature_2 = np.random.normal(400, 15, size=800)
```

```python
feature_3 = np.zeros(800)
for i in range(800):
    if groups[i] == 'A':
        feature_3[i] = 3*feature_1[i] - 5*feature_2[i] + np.random.normal(0, 12)
    elif groups[i] == 'B':
        feature_3[i] = 2*feature_1[i] + np.random.normal(0, 12)
    else:  # Group C
        feature_3[i] = -feature_1[i] + feature_2[i] + np.random.normal(0, 12)
```

Question 1

```python
df = pd.DataFrame({
    'ID': id_series,
    'Group': groups,
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3
})
```

Question 2

```python
theoretical_count = len(feature_1[feature_1 > 3000])

actual_count = len(df[df['Feature_1'] > 3000])

group_stats = df.groupby('Group')['Feature_3'].agg(['mean', 'std'])
```
